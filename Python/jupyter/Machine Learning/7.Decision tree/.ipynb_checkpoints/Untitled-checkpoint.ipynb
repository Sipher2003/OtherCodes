```python
import pandas as pd
```

```python
df=pd.read_csv("titanic.csv")
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis="columns",inplace=True)
df
```

```python
inputdf=df.drop("Survived",axis="columns")
target=df.Survived
```

```python
inputdf
```

```python
target
```

```python
inputdf
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_sex=LabelEncoder()
```

```python
inputdf["Sex_n"]=le_sex.fit_transform(inputdf["Sex"])
inputdf
```

```python
inputdf.Age=inputdf.Age.fillna(inputdf.Age.mean())
inputdf
```

```python
newinputdf=inputdf.drop(["Sex"],axis="columns")
newinputdf
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(newinputdf,target,test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
