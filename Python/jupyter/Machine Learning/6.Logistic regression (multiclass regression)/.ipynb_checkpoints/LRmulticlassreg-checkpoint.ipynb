```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
```

```python
dir(digits)
```

```python
digits.data[0]
```

```python
plt.gray()  #for the grey color
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
digits.target[0:5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(digits.data,digits.target,test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver = "newton-cg", multi_class="auto")
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
plt.matshow(digits.images[44])
```

```python
digits.target[44]
```

```python
model.predict([digits.data[44]])
```

```python
model.predict(digits.data[0:5])
```

```python
#So the model is predicting correctly above. But the score was just 96%. 
#We have to make sure where it went wrong.
#Lets do that
```

```python
ypredicted=model.predict(X_test)#get all the predicted values for te X_test
from sklearn.metrics import confusion_matrix

cm=confusion_matrix(y_test,ypredicted)  #y_test is truth and y_predicted is what the model predicted
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
#above code is just copied to visualuze the array we got above and understand better.
```
