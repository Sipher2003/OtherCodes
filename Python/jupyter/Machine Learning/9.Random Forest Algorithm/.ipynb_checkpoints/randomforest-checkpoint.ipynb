```python
import pandas as pd
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
```

```python
digits=load_digits()
```

```python
dir(digits)
```

```python
plt.gray()
for i in range(4):
    plt.matshow(digits.images[i])
```

```python
df=pd.DataFrame(digits.data)
df.head()
```

```python
df["target"]=digits.target
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop("target",axis="columns")
X
```

```python
y=df["target"]
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=42)
# #The n_estimators hyperparameter controls how many decision trees are included in the random forest.
```

```python
model.fit(X_train,y_train)
```

```python
model.predict()
```

```python
digits.target[31]
```

```python
model.predict(digits.data[[31]])
```

```python
model.score(X_test,y_test)
```

```python
#We have to make sure where it went wrong.
#Lets do that
```

```python
ypredicted=model.predict(X_test)#get all the predicted values for te X_test
from sklearn.metrics import confusion_matrix

cm=confusion_matrix(y_test,ypredicted)  #y_test is truth and y_predicted is what the model predicted
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
#above code is just copied to visualuze the array we got above and understand better.
```
