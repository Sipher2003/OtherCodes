```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
import warnings   #for clean notebook 
```

```python
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("Melbourne_housing_FULL.csv")
df
```

```python
df.nunique()
```

```python
df.shape
```

```python
cols_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 
               'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
df = df[cols_to_use]
df
```

```python
df.isna().sum()
```

```python
# Some feature's missing values can be treated as zero (another class for NA values or absence of that feature)
# like 0 for Propertycount, Bedroom2 will refer to other class of NA values
# like 0 for Car feature will mean that there's no car parking feature with house
cols_to_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
df[cols_to_fill_zero]=df[cols_to_fill_zero].fillna(0)
```

```python
df.isna().sum()
```

```python
df['Landsize']=df['Landsize'].fillna(df['Landsize'].mean())
df['BuildingArea']=df['BuildingArea'].fillna(df['BuildingArea'].mean())
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True) #dropping the 3 rows of regioname and 3 rows of councilarea
df
```

```python
df=pd.get_dummies(df,drop_first=True)
df
```

```python
X=df.drop("Price",axis="columns")
y=df["Price"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=2)
```

```python
reg = LinearRegression().fit(X_train, y_train)
```

```python
reg.score(X_test,y_test)
```

```python
reg.score(X_train,y_train)
```

The score above is very less for test and more for train .This is the case of overfitting.

**Overfitting**: Imagine trying to memorize a story instead of understanding its main ideas. In machine learning, overfitting happens when a model learns the training data too well, including its noise and small fluctuations. This makes the model perform excellently on the training data but poorly on new, unseen data because it's essentially just "memorizing" the training examples.

**Underfitting**: On the other hand, underfitting is like not paying enough attention to the story's details and missing the important points. In machine learning, underfitting occurs when a model is too simple to capture the underlying patterns in the data. This leads to poor performance both on the training data and new data because the model isn't capturing the relevant information.

In essence, overfitting is when a model is too complex, and underfitting is when it's too simple. The goal is to find a balance, where the model learns the essential patterns without getting lost in noise or being too rigid.

Lasso regression is basically L1 regularization.

```python
lasso=linear_model.Lasso(alpha=50,max_iter=100,tol=0.1)
```

```python
lasso.fit(X_train,y_train)
```

```python
lasso.score(X_test,y_test)
```

```python
lasso.score(X_train,y_train)
```

Ridge regression is also called L2 reguarization

```python
from sklearn.linear_model import Ridge
ridge_reg= Ridge(alpha=50, max_iter=100, tol=0.1)
ridge_reg.fit(X_train, y_train)
```

```python
ridge_reg.score(X_test,y_test)
```

```python
ridge_reg.score(X_train,y_train)
```

We see that Lasso and Ridge Regularizations prove to be beneficial when our Simple Linear Regression Model overfits. These results may not be that contrast but significant in most cases.Also that L1 & L2 Regularizations are used in Neural Networks too
