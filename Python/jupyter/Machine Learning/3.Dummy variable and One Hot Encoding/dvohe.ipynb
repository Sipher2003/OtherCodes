```python
import pandas as pd
```

```python
df=pd.read_csv("homeprices.csv")
df
```

```python
dummies=pd.get_dummies(df.town)
dummies
```

```python
merged=pd.concat([df,dummies],axis="columns")
merged
```

```python
final=merged.drop(['town','west windsor'], axis="columns")  #dropping towns column bcoz now we have made dummy variabe of that.            
final
#and dropping west windsor bcoz we it is a good practice to drop one of dummy columns to avoid dummy variable trap.
```

```python
X=final.drop("price",axis="columns")  #dropped price coz thats Y i.e dependent variable
X
```

```python
Y=final.price
Y
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(X,Y)  #training the model
```

```python
model.predict([[2800,0,1]]) #predicting price for 2800 area at robinsville
```

```python
model.predict([[3400,0,0]]) #predicting price for 3400 area at West windsor
```

```python
model.score(X,Y)  #to check how accurate the model is
```
