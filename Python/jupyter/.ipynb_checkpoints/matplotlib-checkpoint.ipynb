```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[23,56,32,21,43,56,27]
```

```python
plt.xlabel("days")
plt.ylabel("temperature")
plt.title("Weather")
plt.plot(x,y,color='green',linewidth=5,ls="dotted")
```

```python
A=[1,2,3,4,5,6,7,8]
B=[23,56,32,21,43,56,27,31]

plt.plot(A,B,"b+")  #means blue colour and + symbol
```

```python
plt.plot(A,B,alpha=0.2)  #alpha basically means visibility
```

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")

plt.legend(loc='lower left',fontsize="small")   #this is to know which line represents what
```
